# Observations

### 1 Temperatures increase closer to the equator
Both plots for the Southern and Northern Hemispheres asssesing temperature against latitude indicated higher temperatures near a latitude of zero or the equator. With temperature dropping off as latitude increases above 0 or moving higher in the northern hemisphere and temperature increasing as the latitude increases to zero in the southern hemisphere. Which is logical due to the suns rays angle of incidence change and being dispersed over a larger surface area as we move away from the equator.However, there was a stronger correlation within the northern hemisphere datum points. This could be due to the northern hemisphere having approximately double the amount of cities with the data frame (366 vs 185).


### 2 Distance from the equator does not dramatically impact wind speeds
Both the Northern and Southern hemisphere displayed very weak correlations when assessing wind speed and latitude (0.0759,-0.2141 respectively). This is unexpected as it would be assumed that the coriolis effect would increase wind speeds at the equator. However, impact could potentially be dampened by the impact of winds moving off of the poles. 

### 3 Humidity increases in the Southern hemisphere towards the equator and inversly decreasing towards the poles in the North.

An interesting pattern occurs when assessing the Northern and Southern hemisphere graphs for humidity vs latitude. In the southern hemisphere we see a slight increase in humidity towards the equator with a very weak positive correlation (0.1422).However, in the Northern hemisphere there is a moderate positive correlation (0.4448) as latitude move towards its upper limit. The trend seen in the southern hemisphere is logical as an increase in temperature would likely result in a higher level of vaporisation and thus, greater levels of humidity. But the northern trend is suprising as the further north the more land there is  and the cooler it becomes.Thus, it is expected that the humidity would be lower. 


```python
# Dependencies
import requests
from api_keys import api_key
import pandas as pd
import sklearn.datasets as dta
import scipy.stats as st
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
!pip install citipy
from citipy import citipy
```

## Generate Cities List

```python
### Generating random long and lat values 
url = "http://api.openweathermap.org/data/2.5/weather?"
filepath = "city.list.json"
bulk_cities = pd.read_json(filepath)


# set seed 
np.random.seed(1)
lat= np.random.randint(-90,90,size = 1500 )
long = np.random.randint(-180,180,size = 1500 )

city_list=[]
for x in range(len(lat)):
    city = citipy.nearest_city(lat[x], long[x])
    city_list.append(city.city_name)
    
city_list= set(city_list)
print(len(city_list))

#city_list
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL

query_url = f"{url}appid={api_key}&units={units}&q="

cities = ["Paris", "London", "Oslo", "Beijing"]

# set up lists to hold reponse info
lat = []
temp = []
lng = []
Max_Temp=[]
Humidity=[]
Cloudiness=[]
Wind_Speed=[]
Country=[]
Date=[] 
names = []
newlist=[]
i = 0
# Loop through the list of cities and perform a request for data on each
for city in city_list:
    # checking for duplicates
    if city in newlist:
        print("Duplicate")
    else:
        newlist.append(city)
        i=i+1
        response = requests.get(query_url + city).json()

        print(f'Processing record {i}: City: {city} ')
        #ames.append(response["name"])
        try:
            names.append(response["name"])
            lat.append(response['coord']['lat'])
            temp.append(response['main']['temp'])
            lng.append(response['coord']['lon'])
            Max_Temp.append(response['main']['temp_max'])
            Humidity.append(response["main"]["humidity"])
            Country.append(response["sys"]["country"])
            Cloudiness.append(response["clouds"]["all"])
            Wind_Speed.append(response["wind"]["speed"])
            Date.append(response["dt"])
        except KeyError:
            print("City not found...Skipping")

#print(f"The latitude information received is: {lat}")
#print(f"The temperature information received is: {temp}")
#print(Country)

city_df = pd.DataFrame(city_list)
```

```python
city_df = pd.DataFrame(names)
city_df["lat"] = lat
city_df["lng"] = lng
city_df["Max_Temp"] = Max_Temp
city_df["Humidity"] = Humidity
city_df["Cloudiness"] = Cloudiness
city_df["Wind_Speed"] = Wind_Speed
city_df["Country"] = Country
city_df["Date"] = Date

city_df.rename(columns = {0:"City"}, inplace = True)

city_df
```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
Northernhemi = clean_city_data[clean_city_data["lat"] > 0 ]
print(len(Northernhemi))

correlation = st.pearsonr(Northernhemi["lat"],Northernhemi["Max_Temp"])

########## Regression
m_slope, m_int, m_r, m_p, m_std_err = st.linregress(Northernhemi["lat"], Northernhemi["Max_Temp"])
print(f'The r-value is: {m_r}')
m_fit = m_slope * Northernhemi["lat"] + m_int

print(f'regression line = {round(m_slope,2)} * Latitude + {round(m_int,2)}')

plt.scatter(Northernhemi["lat"], Northernhemi["Max_Temp"])

plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.plot(Northernhemi["lat"],m_fit, "--",color = "r", label ='y={:.2f}x+{:.2f}'.format(m_slope,m_int))
plt.legend(loc = 4)
plt.grid(True)
plt.savefig("outputs/Northern Hemisphere - Max Temp vs. Latitude Linear Regression.PNG")
```
